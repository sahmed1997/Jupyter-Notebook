### Problem 1: help the pigs

15 points 

You are planning to start a non-profit to rescue abandoned guinea pigs. You've rented a building to convert to a guinea pig rescue center. The building has 800 square feet of space in it. You want to know the maximum number of guinea pigs you can house at your rescue center. You are going to plan cages to hold at most two guinea pigs, and you can dedicate no more than 40% of your space to cages (the rest will be for supplies, visitors, and general comfort). A cage for two guinea pigs should be 10.5 square feet at a minimum.  

If you stick to the minimum cage size, what is the maximum number of guinea pigs you can house? Report two numbers, the number you can house if you put one guinea pig per cage, and the number if you have  two guinea pigs per cage. Be sure to report the final numbers as integers, since you can't house a fraction of a guinea pig. Round down using the floor function from the math module. 

You will write a program to do these calculations. You want your program to be general, in case you rent a different building in the future, or you find out that guinea pigs really need more space, or that a different percentage of the space is needed for cages versus other needs. Therefore your program should be as general as possible.

Use the cell below for your code. Add statements to calculate the two results you need. Use a call to print to display the results, and include string elements to make it more readable.


```python
""""This program calculates the number of guinea pigs that can be housed in a storage depending on total area,
    percent of space for cages, number of pigs per cage and area per cage.

    Author: Salman
"""
import math

buildingSize = 800   # square feet, size of building
cagePercent = 0.40   # percentage dedicated to cages
cageArea = cagePercent * buildingSize #square feet, area available for cages
cageSize = 10.5      # square feet, minimum size of cage for 1-2 guinea pigs
numberCages = math.floor(cageArea / cageSize) #total number of cages
pigsPerCage1 = 1 #capacity of each cage (when 1)
pigsPerCage2 = 2 #capacity of each cage (when 2)
print("When there is one guinea pig per cage,",(numberCages*pigsPerCage1),"pigs can be housed") #print statement when there is 1 pig per cage
print("When there are two guinea pig per cage,",(numberCages*pigsPerCage2),"pigs can be housed") #print statement when there are 2 pigs per cage
```

### Problem 4: If I only had the time...

Python has a [library of functions for working with time][1], including a function called `asctime()`. This function reads the current time from the computer's internal format and returns it in a human-readable format. Devise a program in the cell below that displays the current time and date.

**Optional Challenge:** Make the program display the time every second, or perhaps every 3 or 5 seconds, four times in a row. Looking ahead in your reading, can you determine ways to do this easier?

[1]:https://docs.python.org/3.5/library/time.html

```python
""""This program prints the current time 4 times, with 1 second intervals in between 

    Author: Salman
"""

import time

for a in range(4): #number of iterations = 4
    print(time.asctime()) #prints the current time
    time.sleep(1) #tells the program to pause for 1 (Number in the parenthesis, which could be 3 or 5) second(s)
```
