## Problem: Wind chill calculator

The following information was obtained from [the wikipedia page about wind chill index](https://en.wikipedia.org/wiki/Wind_chill).

The standard wind chill formula for Canada is:

$$T_{\rm wc} = 13.12 + 0.6215T_{\rm a} - 11.37V^{+0.16} + 0.3965T_{\rm a}V^{+0.16}$$

where $T_{\rm wc}$ is the wind chill index, based on the Celsius temperature scale, $T_{\rm a}$ is the air temperature in degrees Celsius (째C), and $V$ is the wind speed at 10 metres (standard anemometer height), in kilometres per hour (km/h)

The standard wind chill formula in the U.S. is:

$$T_{\rm wc}=35.74+0.6215 T_{\rm a}-35.75 V^{+0.16}+0.4275 T_{\rm a} V^{+0.16}\,\!$$

where $T_{\rm wc}$ is the wind chill index, based on the Fahrenheit scale, $T_{\rm a}$ is the air temperature, measured in 째F, and $V$ is the wind speed, in mph.

**Important Note:**
Windchill temperature is defined only for temperatures at or below 10 째C (50 째F) and wind speeds above 4.8 kilometres per hour (3.0 mph). In the functions you will write, you should simply return the given temperature if these conditions are not met.

### Your task
Write separate functions for calculating the Canadian and U.S. wind chill temperature and returning it, given a temperature and wind speed. Then write a main prgram with a function called `test` that tests out each of your functions for various values of temperatures and wind speeds, including those where the important note above applies. 

Once you have your tests working in the `test` function, then add a separate function called `getData` that first asks the user whether they wish to use Celcius or Fahrenheit to compute a wind chill index. Then ask them to provide the temperature and wind speed. Then report the wind chill temperature to the user with the appropriate degrees (print out whether it is in Celcius or Fahrenheit).

```python
'''This program first tests windchill calculations using 3 sets of wind speeds and air temperatures.
    Next, it asks the user if they would like to use the Celsius or Fahrenheit and then uses either the
    Canada formula (for Celsius) or the US formula (for Fahrenheit) to calculate the wind chill index.
 
    Author: Salman Ahmed
'''
    
import random
    
def windChillCanada(at, v):
    '''This function takes 2 inputs, air temperature (at) and wind speed (v). It calculates and returns
       the wind chill index using the Canada formula
    '''
    if at<=10 and v>4.8:
        wc = 13.12+0.6215*at-11.37*(v**0.16)+0.3965*at*(v**0.16)
    else:
        wc = at
    return wc

def windChillUS(at, v):
    '''This function takes 2 inputs, air temperature (at) and wind speed (v). It calculates and returns
       the wind chill index using the US formula
    '''
    if at<=50 and v>3.0:
        wc = 35.74+0.6215*at-35.75*(v**0.16)+0.4275*at*(v**0.16)
    else:
        wc = at
    return wc

def test():
    '''This function takes no inputs and returns nothing. It calls the calculator functions using 3 sets
       of values of air temperature and tests the calculations
    '''
    for a in range(3):
        at = random.randint(0,50)
        v = random.randint(0,10)
        windchill1 = windChillCanada(at, v)
        windchill2 = windChillUS(at, v)
        print("When air temperature is",at ,"degrees celsius and wind speed is", v,"kilometres per hour at 10 metres height, wind chill temperature = ", windchill1,"degrees celsius")
        print("When air temperature is",at ,"degrees fahrenheit and wind speed is", v,"miles per hour at 10 metres height, wind chill temperature = ", windchill2,"degrees fahrenheit") 
    
def getData():
    '''This function takes no inputs and returns nothing. It asks the user if they would like to use the Celsius
       or Fahrenheit and then calls the respective calculator function to print the wind chill temperature
    '''
    choice = int(input("Press 1 if you wish to use Celcius and 2 if you wish to use Fahrenheit to compute a wind chill index"))
    if choice == 1:
        airTemp = float(input("Enter air temperature in degrees celsius"))
        windSpeed = float(input("Enter wind speed in kilometres per hour, at 10 metres height"))
        windChill = windChillCanada(airTemp, windSpeed)
        print("When air temperature is",airTemp ,"and wind speed is", windSpeed,", wind chill temperature = ", windChill,"degrees celsius")
    elif choice == 2:
        airTemp = float(input("Enter air temperature in degrees fahrenheit"))
        windSpeed = float(input("Enter wind speed in miles per hour, at 10 metres height"))
        windChill = windChillUS(airTemp, windSpeed)
        print("When air temperature is",airTemp ,"and wind speed is", windSpeed,", wind chill index = ", windChill,"degrees fahrenheit")
    else:
        print("Invalid entry")
         
def main():
    '''This is the main function which takes no inputs and returns nothing. It calls the test and getData functions
    '''
    test()
    getData()    
        
if __name__ == "__main__":
    main() ##The main function is called here
```
