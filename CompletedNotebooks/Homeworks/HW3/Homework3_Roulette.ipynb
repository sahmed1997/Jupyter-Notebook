```python
""" The program plays American Roulette
    Author: Salman Ahmed (Help taken from Rohit Bagda)
    
"""
from random import randrange 

"""The function returns True if the roulette num is equal to the singleNum chosen by the user and false if otherwise"""
def singleNumPay(singleNum,num):
    if singleNum == num:
        return True
    else:
        return False
"""This function returns true or false based on the user's choice of black or not"""    
def blackPay(num):
    blackList=(2, 4, 6, 8, 10, 11, 13, 15, 17, 20, 22, 24, 26, 28, 29, 31, 33, 35)
    l=len(blackList)
    tmp=0
    for i in range(l-1):
        if blackList[i]==num:
            return True
            tmp=1
    if tmp==0:        
        return False

"""This function returns true or false based on the user's choice of red or not"""
def redPay(num):
    redList =(1, 3, 5, 7, 9, 12, 14, 16, 18, 19, 21, 23, 25, 27, 30, 32, 34, 36)
    l=len(redList)
    tmp = 0
    for i in range(l-1):
        if redList[i]==num:
            return True
            tmp=1
    if tmp==0:        
        return False

"""This function returns true or false based on the user's choice of odd or not"""
def oddPay(num):
    if num%2!=0:
        return True
    else:
        return False

"""This function returns true or false based on the user's choice of even or not"""
def evenPay(num):
    if num%2==0 and num!=0:
        return True
    else:
        return False

"""This function returns true or false based on the user's choice of low or not"""
def lessThan19Pay(num):
    if num<19 and num>0:
        return True
    else:
        return False
    
"""This function returns true or false based on the user's choice of high or not"""    
def greaterThan18Pay(num):
    if num>18 and num<37:
        return True
    else:
        return False

"""This function accepts bet type from the user and calls the respective functions to determine the result"""
def play(): 
    while True:
        num = randrange(0, 38)
       
        betChoice=int(input("""Place a bet:
        1. Single Number
        2. Red or Black
        3. Odd or Even
        4. Low or High
        5. Quit\n"""))
        
        if betChoice == 5: 
            print("Bye!")
            break 

        if betChoice == 1:
            singleNum=input("Enter Number - ")
            if singleNum=="00":
                singleNum = 37
            else:
                singleNum = int(singleNum)
            result=singleNumPay(singleNum,num)
            
            if result == True:
                print("You won the bet")
            else: 
                print("You lost the best ")

        elif betChoice == 2:
            redBlack=input("Enter Red or Black - ")
            if redBlack == "Red":
                result=redPay(num)
            elif redBlack == "Black":
                result=blackPay(num)
            
            if result == True:
                print("You won the bet")
            else: 
                print("You lost the best ")
                
        elif betChoice == 3:
            oddEve=input("Enter Odd or Even - ")
            if oddEve == "Odd":
                result=oddPay(num)
            elif oddEve =="Even":   
                result=evenPay(num)
            
            if result == True:
                print("You won the bet")
                
            else: 
                print("You lost the best ")
                
        elif betChoice == 4:
            highLow = input("Enter High or Low - ")
            if highLow == "Low":
                result= lessThan19Pay(num)
            else:
                result= greaterThan18Pay(num)
            
            if result == True:
                print("You won the bet")
            else: 
                print("You lost the best ")
                
        if num == 37:
            print("Roulette Result - 00")
        else:
            print("Roulette Result - ", num)

    
"""This function tests the pay functions using example values"""        
def test():
    result=singleNumPay(5,5)
    if result==True:
        print("You won the bet")
    else:
        print("You lost the best")

    result=singleNumPay(5,30)
    if result==True:
        print("You won the bet")
    else:
        print("You lost the best")
            
    for i in range(38):
        result=blackPay(i)
        if result==True:
            print(i, "is Black")
        else:
            print(i, "is not Black")
                
    for i in range(38):
        result=redPay(i)
        if result==True:
            print(i, "is Red")
        else:
            print(i, "is not Red")
    
    for i in range(38):
        result=oddPay(i)
        if result==True:
            print(i, "is Odd")
        else:
            print(i, "is not Odd")
            
    for i in range(38):
        result=evenPay(i)
        if result==True:
            print(i, "is Even")
        else:
            print(i, "is not Even")
    
    for i in range(38):
        result=lessThan19Pay(i)
        if result==True:
            print(i, "is Low")
        else:
            print(i, "is not Low")
    
    for i in range(38):
        result=greaterThan18Pay(i)
        if result==True:
            print(i, "is High")
        else:
            print(i, "is not High")
            
"""The main function calls the play and test functions."""    
def main():
    test()
    play()

if __name__ == '__main__':
    main()
```
